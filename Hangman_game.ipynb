## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

```python
import random


def letters_num(word_len):
    for n in range(word_len):
        guess.append('_ ')
    print(''.join(guess))


def secret():
    secret_words_list = ['apple','banana','mango','orange','kiwi','grapes','peach','strawberry',
                    'blueberry','pineapple','watermellon','avocado']

    return random.choice(secret_words_list)


def check_letters(user_guess,secret_word):
    user_guess_list = list(user_guess)
    secret_word_list = list(secret_word)

    user_set = set(user_guess_list)
    sec_set = set(secret_word_list)
    inter = user_set.intersection(sec_set)
    
    for indx , letter in enumerate(secret_word_list):
        if letter in inter:
            guess[indx] = letter
    word = ''.join(guess)
    print(word)
    if word == secret_word:
        print("\nCongratulations! you guessed right!") 
        print(word)
        return True
    return False


# Begining of the Game
while True:
    guess = []
    print("\nWelcome to our Hangman Game!")
    print("Let's play!")
    print("Try to guess the Fruit name.")

    secret_word = secret()
    word_len = len(secret_word)
    letters_num(word_len)

    for chances in range(word_len+2):
        user_guess = input("Enter Fruit name : ")

        if user_guess == secret_word:
            print("\nCongratulations! You guessed right!")
            print(user_guess)
            break
        else:
            if(check_letters(user_guess,secret_word)):
                break
            else:
                if chances == (word_len+1):
                    print("\nOoops! You ran out of chancesðŸ˜”.")
                    print("Good Luck next time.")
    
    print("\nWould you like to play again ? [Y/N]")
    play_again = input("Would you like to play again ? [Y/N]").upper()
    if play_again == 'N':
        print("Thank you for playing Hangman Game!")
        break
```
